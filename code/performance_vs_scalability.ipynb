```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
users = np.linspace(0, 100, 300)

# Stable system (flat line)
scales_well = np.ones_like(users) * 80

# Poorly scaling system (sigmoid-like drop)
scales_poorly = 80 / (1 + np.exp((users - 50) / 7))  # sigmoid curve centered at 50 users

# Create figure (1200x700 pixels)
plt.figure(figsize=(12, 7), dpi=100)

# Plot curves
plt.plot(users, scales_well, linewidth=3, color='orange')
plt.plot(users, scales_poorly, linewidth=3, color='blue')

# Add annotations with arrows and matching text color
plt.annotate(
    "Scales well",
    xy=(70, 80),              # point on the curve
    xytext=(80, 50),          # text position
    arrowprops=dict(facecolor='orange', arrowstyle='->', lw=2),
    fontsize=14,
    color='orange'             # text color matches curve
)

plt.annotate(
    "Scales poorly",
    xy=(50, 40),              # point on the curve
    xytext=(60, 30),          # text position
    arrowprops=dict(facecolor='blue', arrowstyle='->', lw=2),
    fontsize=14,
    color='blue'               # text color matches curve
)

# Labels and title
plt.xlabel("Number of Users", fontsize=14)
plt.ylabel("Performance (in ms)", fontsize=14)
plt.title("System Scalability vs Performance", fontsize=16)

# Styling
plt.grid(True, linestyle="--", alpha=0.5)

# Show plot
# plt.show()

plt.savefig("./../images/performace_vs_scalability.png")
```
